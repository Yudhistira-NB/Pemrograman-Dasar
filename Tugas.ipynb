```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/sample_data/products mentah.csv")
```

```python
df.describe()
```

```python
# Ringkasan statistik
df.describe()

# Rata-rata harga
rata_rata_price = df['price'].mean()
print("Rata-rata harga:", rata_rata_price)

# Modus dari quantity
modus_qty = df['quantity'].mode()
print("Modus size:", modus_qty)
```

```python
df['quantity'].hist()
plt.title("Frekuensi jumlah quantity")
plt.xlabel("Quantity")
plt.ylabel("frekuensi")
plt.show()
```

```python
# Kolerasi
corr_matrix = df[['quantity', 'price']].corr()
print("Korelasi:", corr_matrix)

# Kovarian
cov_matrix = df[['quantity', 'price']].cov()
print("Kovarian:", cov_matrix)
```

```python

KOLOM_GROUPING = 'product_type'

group_mean_data = df.groupby(KOLOM_GROUPING).mean(numeric_only=True)
print(group_mean_data)


group_agg_data = df.groupby(KOLOM_GROUPING).agg({
    'price': ['mean', 'min', 'max'],
    'quantity': ['mean', 'min', 'max']
})
print(group_agg_data)
```
