```python
# Input (Float)
suhu = float(input("Masukkan suhu air pendingin (°C): "))

# Penentuan Status
if suhu <= 90:
    print("STATUS: Normal")
elif suhu < 100:
    print("STATUS: Warning")
else:
    print("ALERT: SHUTDOWN")
```

```python
vibrasi_mesin = [1.2, 0.9, 2.5, 3.1, 1.8]

# Kode Perulangan
for i in range(len(vibrasi_mesin)):
    if vibrasi_mesin[i] > 2.0:
        print(f"Warning: Unit {i + 1} vibrasi tinggi!")
```

```python
suhu_saat_ini = 60

# Loop selama suhu di bawah 85°C
while suhu_saat_ini < 85:
    print(f"Suhu saat ini: {suhu_saat_ini}°C")
    suhu_saat_ini += 5 # Menambah nilai 5 setiap perulangan
```

```python
# Input (Float)
kapasitas_maks = float(input("Masukkan kapasitas maksimum (kW): "))
beban_terpakai = float(input("Masukkan beban terpakai (kW): "))

persentase = (beban_terpakai / kapasitas_maks) * 100

# Pernyataan status
if persentase <= 60:
    status = "Beban Ringan"
elif persentase <= 85:
    status = "Beban Sedang"
else:
    status = "Beban Berat"

# Tampilkan hasil
print(f"Status beban: {status}")
```
