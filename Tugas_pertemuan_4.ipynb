```python
# Modifikasi kode program kasir
print("="*30)
print("SELAMAT DATANG DI KAFE GEMINI")
print("="*30)
print("Menu Makanan & Minuman:")
print("1. Kopi Hitam\t- Rp 12.000")
print("2. Teh Manis\t- Rp 8.000")
print("3. Roti Bakar\t- Rp 15.000")
print("4. Mie Instan\t- Rp 10.000")
print("="*30)

# Inisialisasi variabel untuk menampung total pesanan
list_pesanan = []
total_harga_sebelum_diskon = 0
harga_akhir = 0
diskon_persen = 0
diskon_nilai = 0

# Tambahan fitur loop untuk memungkinkan user memesan lebih dari satu
while True:
    pilihan_str = input("Silakan masukkan nomor pilihan Anda (1-4) atau 'selesai' untuk lanjut pembayaran: ")

    if pilihan_str.lower() == 'selesai':
        break # Keluar dari loop pemesanan

    try:
        pilihan = int(pilihan_str)
    except ValueError:
        print("Input tidak valid. Masukkan nomor menu (1-4) atau 'selesai'.")
        continue

    nama_pesanan = ""
    harga_satuan = 0

    if pilihan == 1:
        nama_pesanan = "Kopi Hitam"
        harga_satuan = 12000
    elif pilihan == 2:
        nama_pesanan = "Teh Manis"
        harga_satuan = 8000
    elif pilihan == 3:
        nama_pesanan = "Roti Bakar"
        harga_satuan = 15000
    elif pilihan == 4:
        nama_pesanan = "Mie Instan"
        harga_satuan = 10000
    else:
        print("Maaf, pilihan menu tidak valid.")
        continue # Lanjut ke iterasi loop berikutnya

    # Tambahan pesanan ke list dan total harga
    list_pesanan.append(nama_pesanan)
    total_harga_sebelum_diskon += harga_satuan
    print(f"Pesanan ditambahkan: {nama_pesanan} (Rp {harga_satuan}). Total sementara: Rp {total_harga_sebelum_diskon}")

# Perhitungan Diskon
print("\n" + "="*30)
print("PROSES PEMBAYARAN")
print(f"Total Belanja Awal: Rp {total_harga_sebelum_diskon}")

# Penentuan diskon berdasarkan skema
if total_harga_sebelum_diskon > 50000:
    diskon_persen = 70
elif total_harga_sebelum_diskon > 20000:
    diskon_persen = 50
elif total_harga_sebelum_diskon > 10000:
    diskon_persen = 30

if diskon_persen > 0:
    diskon_nilai = total_harga_sebelum_diskon * (diskon_persen / 100)
    harga_akhir = total_harga_sebelum_diskon - diskon_nilai
    print(f"Anda mendapatkan diskon {diskon_persen}% (Potongan: Rp {int(diskon_nilai)})")
else:
    harga_akhir = total_harga_sebelum_diskon
    print("Diskon tidak didapatkan. Belanja lebih banyak!")

print(f"TOTAL YANG HARUS DIBAYAR: Rp {int(harga_akhir)}")
print("="*30)

# Proses Pembayaran
if harga_akhir > 0:
    while True: # Loop untuk memastikan user memasukkan uang yang cukup
        try:
            uang_dibayar_str = input(f"Masukkan jumlah uang Anda: Rp ")
            uang_dibayar = int(uang_dibayar_str)
        except ValueError:
            print("Input tidak valid. Masukkan hanya angka.")
            continue

        if uang_dibayar >= harga_akhir:
            kembalian = int(uang_dibayar - harga_akhir)

            # Cetak Struk
            print("\n" + "*"*30)
            print("STRUK PEMBAYARAN")
            print("*"*30)
            print(f"Pesanan Anda\t\t: {', '.join(list_pesanan)}")
            print(f"Total Awal\t\t: Rp {total_harga_sebelum_diskon}")
            print(f"Diskon ({diskon_persen}%)\t\t: Rp -{int(diskon_nilai)}")
            print(f"Total Bayar\t\t: Rp {int(harga_akhir)}")
            print(f"Uang Pembayaran\t\t: Rp {uang_dibayar}")
            print("------------------------------")
            print(f"Kembalian\t\t: Rp {kembalian}")
            print("*"*30)
            print("Terima kasih telah berbelanja!")
            break
        else:
            kekurangan = int(harga_akhir - uang_dibayar)
            print(f"Maaf, uang Anda kurang sebesar Rp {kekurangan}. Mohon masukkan uang lagi.")
            # Program akan meminta input uang_dibayar lagi karena berada di dalam loop while True

print("="*30)
```
