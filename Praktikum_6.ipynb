```python
# SOAL 3: Menghitung Rata-rata dari List Nilai
def hitung_rata_rata(data_nilai):
    """Menghitung rata-rata dari list angka menggunakan perulangan for."""
    total = 0

    # Perulangan untuk menjumlahkan nilai
    for nilai in data_nilai:
        total += nilai

    # Pembagian untuk rata-rata
    if total > 0:
        return total / len(data_nilai)
    return 0

# Pemanggilan:
nilai_praktikan = [85, 78, 92, 65, 90]
print(f"Nilai rata-rata kelas adalah: {hitung_rata_rata(nilai_praktikan)}")
```

```python
def hitung_rata_rata_valid(data_nilai):
    """Menghitung rata-rata hanya dari nilai yang valid (0-100)."""
    total_valid = 0
    jumlah_valid = 0

    # Perulangan dengan pengecekan
    for nilai in data_nilai:
        if (nilai >= 0) and (nilai <= 100):  # Pengecekan apakah nilai berada di rentang 0 sampai 100
            total_valid += nilai
            jumlah_valid += 1
        # Nilai yang < 0 atau > 100 diabaikan (tidak dihitung)

    # Pembagian untuk rata-rata
    if jumlah_valid > 0:
        return total_valid / jumlah_valid
    return 0

# Contoh Pemanggilan dengan Nilai Invalid:
nilai_praktikan_mod = [85, 78, 105, 65, -10, 90, 105]
print(f"\n--- Modifikasi (Mengabaikan Nilai Invalid) ---")
print(f"Nilai rata-rata VALID adalah: {hitung_rata_rata_valid(nilai_praktikan_mod)}")
# 105 dan -10 diabaikan. Rata-rata dihitung dari [85, 78, 65, 90] / 4 = 79.5
```

```python
# SOAL 4: Sistem Penentuan Predikat Nilai
def tentukan_predikat(skor):
    """Menentukan predikat (A, B, C, D, E) berdasarkan skor."""
    if skor >= 85:
        return "A"
    elif skor >= 70:
        return "B"
    elif skor >= 60:
        return "C"
    elif skor >= 50:
        return "D"
    else:
        return "E"

# Pemanggilan:
print(f"Skor 75 mendapat predikat: {tentukan_predikat(75)}")
print(f"Skor 48 mendapat predikat: {tentukan_predikat(48)}")
```

```python
def tentukan_predikat(skor):
    """Menentukan predikat (A, B, C, D, E) berdasarkan skor.
    Memastikan input adalah numerik (int atau float).
    """
    # Modifikasi: Pengecekan Tipe Data
    if not isinstance(skor, (int, float)):
        return "Input harus berupa angka (int/float)"

    if skor >= 85:
        return "A"
    elif skor >= 75:  # Jika skor >= 75 dan < 85 (karena kondisi sebelumnya sudah dicek)
        return "B"
    elif skor >= 60:  # Jika skor >= 60 dan < 75 (karena kondisi sebelumnya sudah dicek)
        return "C"
    elif skor >= 50: # Jika skor >= 50 dan < 60 (karena kondisi sebelumnya sudah dicek)
        return "D"
    else: # Jika skor < 50
        return "E"

# Pemanggilan:
print(f"Skor 75 mendapat predikat: {tentukan_predikat(75)}")
print(f"Skor 48 mendapat predikat: {tentukan_predikat(48)}")
print(f"Skor 85.5 mendapat predikat: {tentukan_predikat(85.5)}")
print(f"Skor 'salah' mendapat predikat: {tentukan_predikat("salah")}")
```

```python
# SOAL 5: Menghitung Faktorial dengan Loop While
def hitung_faktorial(n):
    """Menghitung faktorial dari bilangan positif n."""
    hasil = 1

    # Perulangan berjalan selama n lebih besar dari 1
    while n > 1:
        hasil *= n
        n -= 1  # Kurangi n
    return hasil

# Pemanggilan:
bilangan_n = 4
print(f"Faktorial dari {bilangan_n} adalah: {hitung_faktorial(bilangan_n)}")  # Output: 24
```

```python
# Modifikasi SOAL 5: Menghitung Faktorial dengan Loop While
def hitung_faktorial(n):
    """Menghitung faktorial dari bilangan positif n."""
    if n < 0:
        return "Input tidak valid"
    elif n == 0:
        return 1

    hasil = 1

    # Perulangan berjalan selama n lebih besar dari 1
    while n > 1:
        hasil *= n
        n -= 1  # Kurangi n
    return hasil

# Pemanggilan:
bilangan_n = 4
print(f"Faktorial dari {bilangan_n} adalah: {hitung_faktorial(bilangan_n)}")  # Output: 24
print(f"Faktorial dari 0 adalah: {hitung_faktorial(0)}") # Output: 1
print(f"Faktorial dari -5 adalah: {hitung_faktorial(-5)}")
```

```python
# SOAL 6: Analisis Data Suhu Ekstrem (for dan if/elif)
def analisis_suhu(data_suhu):
    """Menghitung jumlah hari Panas dan Dingin dari data_suhu."""
    panas_count = 0
    dingin_count = 0

    for suhu in data_suhu:
        if suhu >= 35:
            panas_count += 1
        elif suhu <=15:
            dingin_count += 1

    return panas_count, dingin_count

# Pemanggilan:
suhu_mingguan = [36, 30, 14, 25, 38, 16, 28]
jumlah_panas, jumlah_dingin = analisis_suhu(suhu_mingguan)
print(f"Jumlah hari Panas (>=35°C): {jumlah_panas}")
print(f"Jumlah hari Dingin (<=15°C): {jumlah_dingin}")
```

```python
# SOAL 6: Analisis Data Suhu Ekstrem (for dan if/elif)
def analisis_suhu(data_suhu, batas_panas=35, batas_dingin=15):
    """Menghitung jumlah hari Panas dan Dingin dari data_suhu."""
    panas_count = 0
    dingin_count = 0

    for suhu in data_suhu:
        if suhu >= batas_panas:
            panas_count += 1
        elif suhu <= batas_dingin:
            dingin_count += 1

    return panas_count, dingin_count

# Pemanggilan:
suhu_mingguan = [36, 30, 14, 25, 38, 16, 28]
jumlah_panas, jumlah_dingin = analisis_suhu(suhu_mingguan)
print(f"Jumlah hari Panas (>=35°C): {panas_count}")
print(f"Jumlah hari Dingin (<=15°C): {jumlah_dingin}")
```
